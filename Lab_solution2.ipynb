```python
from config import *
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import numpy as np
import pandas as pd


#Initialize SpotiPy with user credentias #
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=client_id, client_secret=client_secret))
```

```python
hot_songs = pd.read_csv('hot_songs.csv') 
hot_songs.shape[0]
```

```python
#=()[]{}
def search_song(title, artist, limit):
    list_of_ids = []
    try:
        results = sp.search(q="tracks: "+title+" artist: "+artist, limit=limit)
        for i in range (0, limit):
            id = results['tracks']['items'][i]['id'] 
            list_of_ids.append(id)
    except:
        print("Song not found!")
        list_of_ids.append("")
    return list_of_ids
```

```python
id = []
X=0
import time
for i in range (0, hot_songs.shape[0]):
    list_of_ids = search_song(hot_songs.iloc[i, 0], hot_songs.iloc[i, 1], 1)
    id.append(list_of_ids[0])
    X=X+1
    print(X)
    time.sleep(5)
    
```
